##**Problem 2.a: Vector Magnitude**<br>
The magnitude of a vector has a given formula which is
$$||v|| = \sqrt{a_0^2 + a_1^2 + ... +a_n^2}$$

But since the dot product of the vector $v$ is equal to the dot product of the same vector $v$
$$v\cdot v = \begin{bmatrix} a_0 \\ a_1 \\ \vdots \\ a_n \end{bmatrix}\cdot \begin{bmatrix} a_0 \\ a_1 \\ \vdots \\ a_n \end{bmatrix}$$

We can say that the dot product of vector $v$ is equal to 
$${a_0^2 + a_1^2 + ... +a_n^2}$$
<br>
Which means that the usual formula $||v|| = \sqrt{a_0^2 + a_1^2 + ... +a_n^2}$ is equal to

$$||v|| = \sqrt{v \cdot v}\\ or \\||v||^2 = v \cdot v$$

##**Problem 2.b: Angle Between Vectors**
Two vectors are $\textrm{orthogonal}$ if the two vectors are *perpendicular* to each other or if they dot product of the two vectors is *zero*.

For example,
$$v = \begin{bmatrix}1 \\ \sqrt2 \\ 1\end{bmatrix}, u = \begin{bmatrix}1 \\ -\sqrt2 \\ 1\end{bmatrix}$$

$$v \cdot u =\begin{bmatrix}1 & \sqrt2 & 1\end{bmatrix} \cdot \begin{bmatrix}1 & -\sqrt2 & 1\end{bmatrix}$$
$$v \cdot u = 1(1) + \sqrt{2}(-\sqrt{2})+1(1)$$
$$v \cdot u = 1 -2 + 1 = 0$$

Since the dot product $v\cdot u$ is equal to 0, the vectors $v$ and $u$ is $\textrm{orthogonal}$. (Perpendicular) <br><br>

Two vectors are $parallel$ if their angle $\theta$ is equal to $0^{\circ}$ or $180^{\circ}$ following the formula $$\theta = cos^{-1}\begin{bmatrix}\frac{ A\cdot B}{||A||\cdot ||B||}\end{bmatrix}$$

For example,
$$v = \begin{bmatrix}4 \\ 3\end{bmatrix}, u = \begin{bmatrix}12 \\ 9\end{bmatrix}$$

$$v \cdot u = 4(12) + 3(9) = 48 + 27 = 75$$

$$||v|| = \sqrt{4^2 + 3^2}= \sqrt{16 + 9} = 5$$

$$||u|| = \sqrt{12^2 + 9^2}= \sqrt{144 + 81} = 15$$

$$\theta = cos^{-1}\begin{bmatrix}\frac{75}{5\cdot 15}\end{bmatrix}$$

$$\theta = cos^{-1}\begin{bmatrix}\frac{75}{75}\end{bmatrix}$$

$$\theta = cos^{-1}(1)$$
$$\theta = 0^{\circ}$$

Since the angle $\theta$ of the two vectors is equal to $0^{\circ}$, the two vectors are $parallel$ to each other.

## **Problem 3.a: Matrix Equivalences**

The matrix multiplication $\theta^T\cdot X$ can also be defined as a table of dot products between all pairs of the two vectors $X$ and $\theta$.

Considering that $X$ and $\theta$ has matrix of:

$$X = \begin{bmatrix} 
— (x^{(1)})^T— \\ 
— (x^{(2)})^T— \\
\vdots \\
— (x^{(m)})^T— \\
\end{bmatrix} \text{, and }
\theta = \begin{bmatrix} 
\theta^{(1)} \\ 
\theta^{(2)} \\
\vdots \\
\theta^{(m)} \\
\end{bmatrix}$$
Since,

$$v^Tw=\begin{bmatrix}v_1 \cdots v_n\end{bmatrix}\begin{bmatrix}w_1 \\ \vdots \\ w_n\end{bmatrix} = v_1w_1+\dots+v_nw_n = v\cdot w$$

Then,
$$\theta^TX = \begin{bmatrix}\theta^{(1)} & \theta^{(2)} & \dots & \theta^{(m)} \\\end{bmatrix} \begin{bmatrix} (x^{(1)})^T \\ (x^{(2)})^T \\ \vdots \\ (x^{(m)})^T \\ \end{bmatrix} = \theta^{(1)}(x^{(1)})^T+\theta^{(2)}(x^{(2)})^T+\dots+\theta^{(m)}(x^{(m)})^T = X \cdot \theta$$

$$\theta^TX=X\cdot\theta$$
